# Spark Quick Test Notebook

This notebook validates a local PySpark setup from CSV download to Parquet output.

## Step 1 - Import PySpark and verify installation

```python
# Import the main PySpark package.
import pyspark
```

```python
# Print the installation path to confirm PySpark is available in this environment.
pyspark.__file__
```

## Step 2 - Create a local Spark session

```python
# Import SparkSession, the entry point for DataFrame operations.
from pyspark.sql import SparkSession
```

```python
# Start a local Spark session using all available cores.
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("spark-local-test") \
    .getOrCreate()
```

## Step 3 - Download and inspect the lookup CSV

```python
# Download the taxi zone lookup file. Use curl for broad OS compatibility.
!curl -fL -o taxi_zone_lookup.csv https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv
```

```python
# Preview the first rows to verify that the file was downloaded correctly.
!head taxi_zone_lookup.csv
```

## Step 4 - Read the CSV file with Spark

```python
# Load the CSV into a Spark DataFrame using the header row.
df = spark.read \
    .option("header", "true") \
    .csv("taxi_zone_lookup.csv")
```

## Step 5 - Write the DataFrame to Parquet

```python
# Persist the DataFrame in Parquet format for efficient analytics.
df.write.mode("overwrite").parquet("zones")
```
